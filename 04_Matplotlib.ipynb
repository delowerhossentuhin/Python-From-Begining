```python
import matplotlib.pyplot as plt
```

```python
x=[1,2,3,4,5,6,7]
y=[50,51,52,48,47,49,46]
plt.xlabel("Day")
plt.ylabel("Temparature")
plt.title("Day VS Temparature")
plt.plot(x,y)
```

```python
plt.plot(x,y,color="green")
```

```python
plt.plot(x,y,color="yEllow",linewidth=3.7)
```

```python
plt.plot(x,y,color="RED",linewidth=3.7,linestyle="dashdot") # 'solid'	'dashed' 'dashdot' 'dotted'
```

```python
plt.plot(x,y,'b+--')
# 'b' = blue color
# '+' = plus marker (each data point marked with a plus sign)
# '--' = dashed line
```

```python
plt.plot(x,y,'b*--')
```

```python
plt.plot(x,y,color="RED",marker='^',linestyle="")
```

```python
plt.plot(x,y,color="RED",marker='^',linestyle="",markersize=20)
```

```python
plt.plot(x,y,'g<',alpha=0.5) # alpha can be specified on a scale 0 to 1
```

```python
import numpy as np
x=np.arange(-20,20,0.1)
x1=np.sin(x)
y1=np.cos(x)
plt.figure(figsize=(10,7))
plt.plot(x,x1)
plt.plot(x,y1)
plt.legend(["sin",'cos'],loc="upper right")
plt.show()
```

```python
plt.plot(x,x1,x,y1)
plt.legend(["sin",'cos'],loc="upper right")
plt.show()
```

```python
plt.figure(figsize=(20,5))
plt.subplot(1,2,1)
plt.plot(x,x1)
plt.legend(["sin"],loc="upper right")
plt.subplot(1,2,2)
plt.plot(x,y1)
plt.legend(['cos'],loc="upper right")
plt.show()
```

```python
from IPython import display
```

```python
sin = []
cos = []
x = []
plt.figure(figsize=(20,7))
for i in range(-50,50,1):
  x.append(i)
  sin.append(np.sin(i))
  cos.append(np.cos(i))
  plt.subplot(1,2,1)
  display.clear_output(wait=True)
  plt.plot(x,sin)
  plt.subplot(1,2,2)
  display.clear_output(wait=True)
  plt.plot(x,cos)
  plt.show()
```

```python
x=np.arange(-20,20,0.1)
x1=np.sin(x)
y1=np.cos(x)
plt.figure(figsize=(20,5))
plt.subplot(1,2,1)
plt.plot(x,x1)
plt.legend(["sin"],loc="upper right")
plt.subplot(1,2,2)
plt.plot(x,y1)
plt.legend(['cos'],loc="upper right")
plt.savefig("Test.pdf",transparent=True)
plt.show()
```
